```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title("PyBer Fare By Month")

plt.ylim(0, 45)

plt.legend()
plt.grid()
```

```python
plt.bar(x_axis, y_axis, color='green', label='boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")

plt.legend()
```

```python
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.invert_yaxis()
```

```python
plt.scatter(y_axis, x_axis, color='r')
plt.gca().invert_yaxis()

plt.xlim(0, 50)
```

```python
plt.scatter(x_axis, y_axis, s=[point * 3 for point in y_axis], edgecolor='black')
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values)
plt.show()
```
